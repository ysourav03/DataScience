# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Reading CSV file

```python
df = pd.read_csv('zomato.csv')
df
```

```python
df.shape
```

```python
df.columns
```

# Dropping columns which are not important

```python
df = df.drop(['url' , 'address' , 'phone', 'menu_item' , 'dish_liked' , 'reviews_list'] , axis = 1)
df
```

```python
df.info() 
```

# Data cleaning

```python
# dropping duplicate row
df.drop_duplicates(inplace=True)
df.shape
```

# Cleaning rate column

```python
df['rate'].unique()
```

# Removing "NEW" , "-" and "/5" from rate column

```python
def removeOtherData(value):
    if (value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

    
df['rate'] =df['rate'].apply(removeOtherData)
df['rate'].head()
        
```

```python
#finding the how many null values are their in rate column
df.rate.isnull().sum()
```

# Filling null value in rate column with mean

```python
# we filling the null values with mean value of rate column
df['rate'].fillna(df['rate'].mean(), inplace = True)
df.rate.isnull().sum()
```

```python
df.info()
```

# Dropping null values

```python

df.dropna(inplace = True)
df.head()
```

```python
df.info()
```

```python
#Renaming the column approx_cost(for two people) and  listed_in(type) 
df.rename( columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'} , inplace =True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

# Listed in(city) and location, both are there, lets keep only one

```python
df.drop(['listed_in(city)'] , axis =1)
```

```python
df['Cost2plates'].unique()
```

# Removing ',' from Cost2Plates Column

```python
def removed(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' ,'')
        return float(value)
    else:
        return float(value)
df['Cost2plates'] = df['Cost2plates'].apply(removed)
```

```python
df['Cost2plates'].unique()
```

```python
df.head()
```

# Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making rest types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

# Cleaning location column

```python
location = df['location'].value_counts(ascending = False)
location
```

```python
locationlessthan300 = location[location < 300]
locationlessthan300
```

# Making location less than 300 in frequency as others

```python
def handle_location(value):
    if value in locationlessthan300:
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning cuisines column

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisineslessthan100 = cuisines[cuisines < 300]
cuisineslessthan100
```

# Making cuisines less than 100 in frequency as others

```python
def handle_cuisiness(value):
    if value in cuisineslessthan100:
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisiness)
df['cuisines'].value_counts()
```

```python
df.head()
```
